```python
import matplotlib.pyplot as plt 
import numpy as np 

losses = [0.03587371313146182, 0.03006847823659579, 0.04579777393844866, 0.0539491848221847, 0.011280]
loss_stds = [0.011603276403369833, 0.008139763704176592, 0.011763500576577265, 0.015171070929939866, 0.026868]
labels = ['MLP VAE', 'MLP VQVAE', 'CNN VAE', 'CNN VQVAE', "Contrastive AutoEncoder"]

# 2. Setup Plot
bar_colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#D6A8FF']

# 3. Setup Plot
plt.figure(figsize=(10, 6)) # Making the chart wider helps prevent overlap
x_pos = np.arange(len(labels))

# 4. Create Bars with custom Width and Colors
# width: Set to 0.6 (default is 0.8) to make bars thinner and increase spacing
plt.bar(x_pos, losses, 
        yerr=loss_stds, 
        capsize=10, 
        color=bar_colors, 
        edgecolor='black', 
        width=0.6)

# 5. Labels & Title
plt.xticks(x_pos, labels)

# 4. Add Labels and Title
plt.ylabel('Average Loss')
plt.title('Reconstruction Loss on Test Set for Three Block Stacking')

# 5. Show/Save
plt.show()
```
